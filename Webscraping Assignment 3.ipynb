```python
from bs4 import BeautifulSoup
import requests
response=requests.get("https://www.flipkart.com/search?q=smartphone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&as-pos=2&as-type=RECENT&suggestionId=smartphone%7CMobiles&requestId=c5cd16e1-ba1b-458d-a3c6-97c16ca2176a&as-searchtext=smart")
soup=BeautifulSoup(response.content,'html.parser')
```

```python
mobile=[]
ram_rom=[]
camera=[]
processor=[]
battery=[]
price=[]
product_url=[]
```

```python
mobiles=soup.find_all('div',class_="_4rR01T")
for i in mobiles:
    mobile.append(i.text)
```

```python
ra=soup.find_all('ul',class_='_1xgFaf')
```

```python
size=[]
for k in ra:
    r=k.find('li',attrs={'class':'rgWa7D'}).text.strip()
    ram_rom.append(ram)
    
    mi=k.find_all('li',attrs={'class':'rgWa7D'})[1].text
    size.append(mi)
    
    my=k.find_all('li',attrs={'class':'rgWa7D'})[2].text
    camera.append(my)
    
    md=k.find_all('li',attrs={'class':'rgWa7D'})[3].text
    battery.append(md)
    
    me=k.find_all('li',attrs={'class':'rgWa7D'})[4].text
    processor.append(me)
```

```python
url=soup.find_all('a',class_='_1fQZEK')
for i in url:
    product_url.append(i['href'])
```

```python
import pandas as pd 
smart=pd.DataFrame({})
smart['Product Name and Brand']=mobile
smart['RAM and ROM']=ram_rom
smart['Display sze and Resolution']=size
smart['Front and Rear Camera']=camera
smart['Battery Capacity']=battery
smart['Processor']=processor
smart['Product Url']=product_url
smart
```

```python
import requests
from bs4 import BeautifulSoup
response=requests.get("https://trak.in/india-startup-funding-investment-2015/")
soup=BeautifulSoup(response.content,'html.parser')
```

```python
data=soup.find_all('tbody',class_='row-hover')[9]
```

```python
date=[]
industry=[]
sub_industry=[]
city=[]
investor_name=[]
investment_type=[]
amount=[]
```

```python
da=data.find_all('td',class_='column-2')
for i in da:
    date.append(i.text)
```

```python
ind=data.find_all('td',class_='column-4')
for l in ind:
    industry.append(l.text)
```

```python
sub=data.find_all('td',class_='column-5')
for m in sub:
    sub_industry.append(m.text)
```

```python
cit=data.find_all('td',class_='column-6')
for n in cit:
    city.append(n.text)
```

```python
inves=data.find_all('td',class_='column-7')
for p in inves:
    investor_name.append(p.text)
```

```python
invest=data.find_all('td',class_='column-8')
for q in invest:
    investment_type.append(q.text)
```

```python
investment=data.find_all('td',class_='column-9')
for r in investment:
    amount.append(r.text.replace('\n',''))
```

```python
import pandas as pd 
startup=pd.DataFrame({})
startup['Date']=date
startup['Industry']=industry
startup['Sub Vertical']=sub_industry
startup['City']=city
startup['Investor_name']=investor_name
startup['Investment_type']=investment_type
startup['Amount (In USD)']=amount
startup
```

```python
import requests
from bs4 import BeautifulSoup
response=requests.get("https://www.digit.in/top-products/best-gaming-laptops-40.html")
soup=BeautifulSoup(response.content,'html.parser')
```

```python
brand=[]
brands=soup.find_all('div',class_='TopNumbeHeading active sticky-footer')
for i in brands:
    name=i.a.h3
    brand.append(name.text)
```

```python
ra=soup.find_all('div',class_='Specs-Wrap')
```

```python
os=[]
display=[]
processor=[]
memory=[]
```

```python
for a in ra:
    name=a.find_all('div',class_='value')
    me=name[0].text
    os.append(me)
    
    my=name[1].text
    display.append(my)
    
    mi=name[2].text
    processor.append(mi)
    
    mn=name[3].text
    memory.append(mn)
```
