```python
import pandas as pd
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
driver=webdriver.Chrome("C:\\Users\\batha\\Downloads\\chromedriver_win32\\chromedriver.exe")
```

```python
title=[]
rating=[]
price=[]
```

```python
driver.get("https://www.amazon.in/s?k=laptops&i=computers&rh=n%3A1375424031%2Cp_n_feature_thirteen_browse-bin%3A12598163031%7C16757432031&dc&qid=1617877917&rnid=12598141031&ref=sr_nr_p_n_feature_thirteen_browse-bin_16")
```

```python
for d in driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']"):
    title.append(d.text)
for e in driver.find_elements_by_xpath("//span[@class='a-price-whole']"):
    price.append(e.text)
```

```python
#locating rating
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
ur=[]
for i in urls[:10]:
    ur.append(i.get_attribute('href'))
for url in ur:
    driver.get(url)
    try:
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()
        ratings=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        rating.append(ratings.text)
        
    except NoSuchElementException  as e:
        rating.append("NO Rating")
```

```python
laptop=pd.DataFrame({})
laptop["Title"]=title[0:10]
laptop["Price"]=price[0:10]
laptop["Rating"]=rating
laptop
```
