# Handle Nan values

first lets handle categorical features which are missing

```python
cat_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='object']
for feature in cat_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

replacing missing values with new label

```python
def replace_cat_nan(df,cat_nan):
    data=df.copy()
    data[cat_nan]=data[cat_nan].fillna('Missing')
    return data

df=replace_cat_nan(df,cat_nan)
df[cat_nan].isnull().sum()
```

```python
df.head()
```

lets handle numerical features which are missing

```python
num_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='object']
for feature in num_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
for feature in num_nan:
    median_value=df[feature].median()
    
    #create a new feature to capture nan values
    df[feature+'nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[num_nan].isnull().sum()
```

```python
df.head(10)
```

some new feature have been added in the dataset like LotFrontagenan,MasVnrAreanan,GarageYrBltnan

```python
#temporal variables (date time varables)
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df['YrSold']-df[feature]
df.head()
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

we have changed YearBuilt', 'YearRemodAdd', 'GarageYrBlt in numercal values

# Numerical variables

Since the numerical variables are skewed we will perform log normal distribution

```python
df.head()
```

```python
import numpy as np
numerical_features=['LotFrontage', 'LotArea', '1stFlrSF','GrLivArea','SalePrice']

for feature in numerical_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

# Handling rare categorical features
we will remove categorical variables that are present less than 1% of the observation

```python
for feature in cat_col:
    rare=df.groupby(feature)['SalePrice'].count()/len(df)
    rare_df=rare[rare>0.01].index
    df[feature]=np.where(df[feature].isin(rare_df),df[feature],'Rare_var')
df.head()
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for feature in cat_col:
    df[feature]=le.fit_transform(df[feature])
```

# Model selection

# Input split

```python
x=df1.drop(columns=['Id','SalePrice'],axis=1)
y=df1['SalePrice']
```

# spliting the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.model_selection import cross_val_score
import numpy as np
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
predlr=lr.predict(x_test)
print('Accuracy score :',lr.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(lr,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predlr))
print('Mean squared error:',mean_squared_error(y_test,predlr))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predlr)))
```

# Lasso regression

```python
from sklearn.linear_model import Lasso,Ridge
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
predls=ls.predict(x_test)
print('Accuracy score :',ls.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(ls,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predls))
print('Mean squared error:',mean_squared_error(y_test,predls))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predls)))
```

# Ridge regression

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
predrd=rd.predict(x_test)
print('Accuracy score :',rd.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(rd,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predrd))
print('Mean squared error:',mean_squared_error(y_test,predrd))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predrd)))
```

# ElasticNet

```python
from sklearn.linear_model import ElasticNet
enr=ElasticNet(alpha=0.0001)
enr.fit(x_train,y_train)
predenr=enr.predict(x_test)
print('Accuracy score :',enr.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(enr,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predenr))
print('Mean squared error:',mean_squared_error(y_test,predenr))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predenr)))
```

# SVR

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
svr.fit(x_train,y_train)
predsvr=svr.predict(x_test)
print('Accuracy score :',svr.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(svr,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predsvr))
print('Mean squared error:',mean_squared_error(y_test,predsvr))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predsvr)))
```

# RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
print('Accuracy score :',rfr.score(x_train,y_train))
print('\n Cross validation score :',cross_val_score(rfr,x,y,cv=8).mean())
print('\n error')
print('Mean absolute error:',mean_absolute_error(y_test,predrfr))
print('Mean squared error:',mean_squared_error(y_test,predrfr))
print('Root Mean squared Error:',np.sqrt(mean_squared_error(y_test,predrfr)))
```
